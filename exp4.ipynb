```python
import numpy as np
def s(x):
    return 1 / (1 + np.exp(-x))
def sd(y):
    return y * (1 - y)
def r(x):
    return np.maximum(0, x)
def rd(x):
    return (x > 0).astype(float)
X = np.array([[1, 1, 0, 1]])
Y = np.array([[1]])
l = 0.69
t = 0.001
W = np.array([
    [ 0.3,  0.1],
    [-0.2,  0.4],
    [ 0.2, -0.3],
    [ 0.1,  0.4]
])
B = np.array([[0.2, 0.1]])
V = np.array([
    [-0.3],
    [ 0.2]
])
C = np.array([[-0.3]])
e = 0
p = False
while True:
    e += 1
    z = np.dot(X, W) + B
    h = r(z)
    u = np.dot(h, V) + C
    o = s(u)
    d = Y - o
    do = d * sd(o)
    dh = do.dot(V.T) * rd(z)
    V += l * h.T.dot(do)
    C += l * do
    W += l * X.T.dot(dh)
    B += l * dh
    if not p:
        print("\n First Epoch")
        print("Output :", o)
        print("Error :", d)
        print("After updating W1:\n", W)
        print("After updating b1:\n", B)
        print("After updating W2:\n", V)
        print("After updating b2:\n", C)
        p = True
    if abs(d[0][0]) < t:
        print("\n Last Epoch")
        print("Epoch:", e)
        print("Output :", o)
        print("Error:", d)
        print("Final W1:\n", W)
        print("Final b1:\n", B)
        print("Final W2:\n", V)
        print("Final b2:\n", C)
        break
```

```python
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def sigmoid_derivative(y):
    return y * (1 - y)
def relu(x):
    return np.maximum(0, x)
def relu_derivative(x):
    return (x > 0).astype(float)
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)
def tanh(x):
    return np.tanh(x)
def tanh_derivative(x):
    return 1 - np.tanh(x)**2
def swish(x):
    return x / (1 + np.exp(-x))
def swish_derivative(x):
    sig = 1 / (1 + np.exp(-x))
    return sig + x * sig * (1 - sig)
def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)
def softmax_derivative(softmax_output):
    s = softmax_output.reshape(-1, 1)
    return np.diagflat(s) - np.dot(s, s.T)
X = np.array([[1, 1, 0, 1]])
Y = np.array([[1]])   
l = 0.69
t = 0.001
W = np.array([
    [0.3, 0.1],
    [-0.2, 0.4],
    [0.2, -0.3],
    [0.1, 0.4]
])
B = np.array([[0.2, 0.1]])
V = np.array([
    [-0.3],
    [0.2]
])
C = np.array([[-0.3]])
e = 0
p = False
while True:
    e += 1

    # --- Forward Pass ---
    z = np.dot(X, W) + B
    h = relu(z)   # you can swap with leaky_relu, tanh, swish
    u = np.dot(h, V) + C
    o = sigmoid(u)   # swap with softmax for multi-class

    # --- Error ---
    d = Y - o

    # --- Backpropagation ---
    do = d * sigmoid_derivative(o)
    dh = do.dot(V.T) * relu_derivative(z)

    # --- Update Weights ---
    V += l * h.T.dot(do)
    C += l * do
    W += l * X.T.dot(dh)
    B += l * dh

    # --- Print First Epoch ---
    if not p:
        print("\n--- First Epoch ---")
        print("Output:", o.item())
        print("Error:", d.item())
        print("Updated W1:")
        for i in range(W.shape[0]):
            for j in range(W.shape[1]):
                print(f"  W{i+1}{j+1}: {W[i,j]}")
        print("Updated b1:")
        for j in range(B.shape[1]):
            print(f"  b1_{j+1}: {B[0,j]}")
        print("Updated W2:")
        for i in range(V.shape[0]):
            print(f"  W2_{i+1}: {V[i,0]}")
        print("Updated b2:")
        print(f"  b2: {C[0,0]}")
        p = True

    # --- Stop Condition ---
    if abs(d.item()) < t:
        print("\n--- Last Epoch ---")
        print("Epoch:", e)
        print("Output:", o.item())
        print("Error:", d.item())
        print("Final W1:")
        for i in range(W.shape[0]):
            for j in range(W.shape[1]):
                print(f"  W{i+1}{j+1}: {W[i,j]}")
        print("Final b1:")
        for j in range(B.shape[1]):
            print(f"  b1_{j+1}: {B[0,j]}")
        print("Final W2:")
        for i in range(V.shape[0]):
            print(f"  W2_{i+1}: {V[i,0]}")
        print("Final b2:")
        print(f"  b2: {C[0,0]}")
        break
```
